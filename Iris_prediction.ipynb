```python
import numpy as np
import pandas as pd
import matplotlib as mlt
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn import svm
```

```python
dataset=pd.read_csv('/content/iris.data',header=None)
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset[4].value_counts()
```

```python
encoder=LabelEncoder()
lables=encoder.fit_transform(dataset[4])
dataset['target']=lables
```

```python
dataset
```

```python
# iris-setosa --> 0
# iris-versicolor --> 1
# iris-virginica --> 2

dataset['target'].value_counts()
```

```python
dataset.drop([4],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
X=dataset.drop(['target'],axis=1)
Y=dataset['target']
```

```python
X
```

```python
Y
```

```python
X.std()
```

```python
scaler=StandardScaler()
standard_X=scaler.fit_transform(X)
```

```python
standard_X
```

```python
standard_X.std()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(standard_X,Y,test_size=0.1,random_state=4)
```

```python
print(X_train.shape, X_test.shape, X.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
training_data_prediction=model.predict(X_train)
accuracy_on_training_data=accuracy_score(training_data_prediction,Y_train)
print("the accuracy of training data: ", accuracy_on_training_data)
```

```python
testing_data_prediction=model.predict(X_test)
accuracy_on_testing_data=accuracy_score(testing_data_prediction,Y_test)
print("the accuracy of testing data: ", accuracy_on_testing_data)
```

```python
model2=svm.SVC(kernel='linear')
```

```python
model2.fit(X_train,Y_train)
```

```python
predict_training_svm=model2.predict(X_train)
accuracy_training_svm=accuracy_score(predict_training_svm,Y_train)
print("the accuracy on training data through svm is: ", accuracy_training_svm)
```

```python
predict_testing_svm=model2.predict(X_test)
accuracy_testing_svm=accuracy_score(predict_testing_svm,Y_test)
print("the accuracy on testing data through svm is: ", accuracy_testing_svm)
```
